```python
import cv2
import numpy as np
from math import sqrt
import os

name = "org_08.jpg"

# Width of image you want to work with
W = 850.
# get original image
oriimg =cv2.imread('assets/'+name,1)
height, width, depth = oriimg.shape
# rescale to W width
imgScale = W/width
newX,newY = oriimg.shape[1]*imgScale, oriimg.shape[0]*imgScale
frame = cv2.resize(oriimg,(int(newX),int(newY)))

# Convert BGR to HSV
hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
# define range of red color in HSV
lower_blue = np.array([0,80,190])
upper_blue = np.array([180,255,255])
# Threshold the HSV image to get only red colors
mask = cv2.inRange(hsv, lower_blue, upper_blue)

kernel = np.ones((3,3),np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

# Bitwise-AND mask and cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),5)original image
res = cv2.bitwise_and(frame,frame, mask= mask)

# contour around mask
i, contours, heirarchy = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# cont_sorted contains the contours    
cont_sorted = sorted(contours, key=cv2.contourArea, reverse=True)[:]

print(len(cont_sorted))

_p=[]
#threshhold is 10 pixels for a 500 pixel wide image. scale accordingly
width_thresh=10/500*W 
height_thresh=10/500*W

# iterate through the contours
for idx, c in enumerate(cont_sorted):
#     get the bounding boxes
    x, y, w, h = cv2.boundingRect(c)
#     only consider bounding boxes larger than thresh
    if(w > width_thresh or h > height_thresh):                
        _p.append([x,y,w,h])
#   _p now contains the bounding boxes of large contours
    cv2.rectangle(res,(x, y),(x+w,y+h),(0,255,0),1)

helpful = 1

while helpful == 1:
    helpful = 0
    print(len(_p))
    
    # __p will contain the bounding boxes of merged contours i.e. one box per schematic symbol
    __p=[]
    
    # iterate through each bounding box
    for idx_1, p_1 in enumerate(_p):
        done =0
    #     check all other bounding boxes for a box that's close to our box
        for idx_2, p_2 in enumerate(_p):
            x_1 = p_1[0]
            x_2 = p_2[0]
            y_1 = p_1[1]
            y_2 = p_2[1]
            w_1 = p_1[2]
            w_2 = p_2[2]
            h_1 = p_1[3]
            h_2 = p_2[3]
    #         if the bounding boxes are close together, merge them into one and put in __p
#      multiplier should be 1.3 or more
            if idx_1 != idx_2 and sqrt((x_1+w_1/2-(x_2+w_2/2))**2 + ((y_1+h_1/2)-(y_2+h_2/2))**2) < max(w_1, w_2, h_1, h_2)*1:
                __p.append([
                    int(min(x_1, x_2)),
                    int(min(y_1, y_2)),
                    int((max(x_1+w_1, x_2+w_2)-min(x_1, x_2))),
                    int((max(y_1+h_1, y_2+h_2)-min(y_1, y_2)))                
                ])
                done =1
                helpful = 1
    #       didn't find a bounding box close to ours, so just add it to the __p list
        if done == 0:
            __p.append(p_1)
    
    if helpful == 1:
        seen = set()
        newlist = []
        for item in __p:
            t = tuple(item)
            if t not in seen:
                newlist.append(item)
                seen.add(t)
        _p = seen

        
# remove duplicates from __p 
# https://stackoverflow.com/questions/21065347/remove-duplicates-in-two-dimensional-array-while-maintaining-sequence
seen = set()
newlist = []
for item in __p:
    t = tuple(item)
    if t not in seen:
        newlist.append(item)
        seen.add(t)
        
# destroy giant boxes        
average_width = 0        
for p in seen:
    average_width = average_width + p[2]/len(seen)
print(average_width)

___p=[]

for p in seen:
    if (p[2] - average_width) < average_width *0.5:
        ___p.append(p)
        
print(len(___p)) # should be 70
        
# iterate through all the bounding boxes for schematic symbols
for idx, p in enumerate(___p):
    x = p[0]
    y = p[1]
    w = p[2]
    h = p[3]
    
    H = max(w, h)
    W = H
    
    x_center = x+w/2
    y_center = y+h/2
    
    X = int(x_center - W/2)
    Y = int(y_center - H/2)
    
    roi_alpha = 0
    
    roi = frame[int(Y-H*roi_alpha):int(Y+(1+roi_alpha)*H), int(X-W*roi_alpha):int(X+(1+roi_alpha)*W)]
    
    newpath = "./assets/" + name.split('.')[0]
    if not os.path.exists(newpath):
        os.makedirs(newpath)
    
    cv2.imwrite("./assets/" + name.split('.')[0] + "/" + str(idx) + ".png", roi)
    
    cv2.rectangle(frame,(X, Y),(X+W,Y+H),(0,255,0),1)
    
        
cv2.imshow('mask',mask)
cv2.imshow('res',res)
cv2.imshow('frame',frame)

k = cv2.waitKey(0) & 0xFF

if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('messigray.png',img)
    cv2.destroyAllWindows()
```

```python
# Standard imports
import cv2
import numpy as np;
 
# Read image
im = cv2.imread("assets/org_06.jpg", cv2.IMREAD_GRAYSCALE)
 
# Set up the detector with default parameters.
detector = cv2.SimpleBlobDetector()
 
# Detect blobs.
keypoints = detector.detect(im)
 
# Draw detected blobs as red circles.
# cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS ensures the size of the circle corresponds to the size of blob
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
 
# Show keypoints
cv2.imshow("Keypoints", im_with_keypoints)
cv2.waitKey(0)
```
